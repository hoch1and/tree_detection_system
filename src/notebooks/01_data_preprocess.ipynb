### Обработка аннотаций (разметки) из директории `/train`

```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/raw_dataset.csv')
print(df.shape[0])
df.sample(3, random_state=42)
```

___

```python
print(f'shape before preprocessing: {df.shape}')
preprocessed_df = df[['file_name', 'bbox', 'segmentation', 'category_id']]

preprocessed_df['bbox'] = preprocessed_df['bbox'].apply(lambda row: [round(float(value)) for value in eval(row)])
preprocessed_df['segmentation'] = preprocessed_df['segmentation'].apply(lambda lst: [round(float(value)) for value in eval(lst)[0]])

display(preprocessed_df.sample(3, random_state=42))
print(f'shape after preprocessing: {preprocessed_df.shape}')
```

```python
print('data info')
display(preprocessed_df.info())
print('\nnull values check')
display(preprocessed_df.isna().sum())
```

```python
preprocessed_df.to_pickle('../data/dataset.pkl')
```
